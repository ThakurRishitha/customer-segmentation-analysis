```python
snapshot_date = df["Purchase Date"].max() + pd.Timedelta(days=1)

print(snapshot_date)
```

```python
rfm = df.groupby("Customer ID").agg({
    "Purchase Date": lambda x: (snapshot_date - x.max()).days,
    "Customer ID": "count",
    "Total Purchase Amount": "sum"
})

rfm.columns = ["Recency", "Frequency", "Monetary"]

rfm.head()
```

```python
print(rfm.shape)

rfm.describe()
```

Choose the value of K where the curve starts to bend (the "elbow"). For this project, we will use **4 clusters**.

```python
kmeans = KMeans(
    n_clusters=4,
    random_state=42
)

rfm["Cluster"] = kmeans.fit_predict(rfm_scaled)

rfm.head()
```

```python
rfm["Cluster"].value_counts()
```

```python
cluster_summary = rfm.groupby("Cluster").mean()

cluster_summary
```

```python
rfm["Cluster"].value_counts().plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Customers in Each Cluster")

plt.xlabel("Cluster")

plt.ylabel("Number of Customers")

plt.show()
```
